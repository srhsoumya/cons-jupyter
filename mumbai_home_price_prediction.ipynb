# we have got this mumbai dataset from kaggle 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Mumbai1.csv",index_col=0)
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.Price=df.Price.apply(lambda x:x/100000)
```

```python
df
```

```python
df.describe()
```

```python
loc=df.groupby('Location').Location.agg('count').sort_values(ascending=False)
loc_count=loc[loc>16]
loc_count

```

```python
df.Location=df.Location.apply(lambda x:x if x in loc_count else 'others')
```

```python
df
```

```python
df=df[df.Area/df["No. of Bedrooms"]>=300]
```

```python
df
```

```python
df_cor=df.corr()
df_cor
```

```python
plt.axis("off")
plt.figure(figsize=(16,16))
sns.heatmap(df_cor,annot=True)
```


# all price depends on is area and no of bedrooms

```python
dfnew=df[["Price","Area","Location","No. of Bedrooms"]]
```

```python
dfnew
```

```python
meani=dfnew.Price.mean()
stdi=dfnew.Price.std()
meani,stdi
```

```python
dfnew=dfnew[dfnew.Price>(stdi-meani)]
dfnew=dfnew[dfnew.Price<=(meani+stdi)]
dfnew
```

```python
dummies=pd.get_dummies(dfnew.Location)
```

```python
dummies
```

```python
finaldf=pd.concat([dfnew,dummies],axis=1)
finaldf
```

```python
finaldf.drop(["Location"],axis=1,inplace=True)
finaldf
x=finaldf.drop(["Price"],axis=1)
y=finaldf.Price
x
```

```python
from sklearn.linear_model import Ridge,Lasso,LinearRegression
mod1=Ridge()
mod2=Lasso()
mod3=LinearRegression()
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y)
```

```python
mod1.fit(xtrain,ytrain)
mod2.fit(xtrain,ytrain)
mod3.fit(xtrain,ytrain)
mod1.score(xtrain,ytrain),mod1.score(xtest,ytest),mod2.score(xtrain,ytrain),mod2.score(xtest,ytest),mod3.score(xtrain,ytrain),mod3.score(xtest,ytest)
```

```python
import pickle 
with open('mumbai_modelfinal.pickle','wb') as file:
    pickle.dump(mod3,file)
```
