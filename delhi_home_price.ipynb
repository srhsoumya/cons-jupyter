```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("MagicBricks.csv")
df
```

```python
df=df[["Area","BHK","Bathroom","Locality","Price"]]
```

```python
df=df[df.Area>=700]
df
```

```python
df.describe()
```

```python
df=df[df.Area<=5000]
```

```python
df.describe()
```

```python
df=df[df.Area/df.BHK>=300]
df
```

```python
df=df[df.BHK+1>=df.Bathroom]
df
```

```python
df.Price=df.Price.apply(lambda x:x/100000)
df
```

```python
df.Locality=df.Locality.apply(lambda x:'others' if len(x.split())>4 else x)
```

```python
loc=df.groupby('Locality').Locality.agg('count').sort_values(ascending=False)
loc=loc[loc>=4]
loc
```

```python
def correct(x):
    splits=x.split()
    lis=['Rohini','Dwarka','Lajpat','Budh','Kailash','Shahdara','Okhla','Greater Kailash','Karol',"Uttam","Vasant",'Patel','Chandini','Saket',"Safdarjung",'Paschim','Alaknanda',"Sarita",'Sultanpur','Chhattarpur','Manglapuri','Mehrauli', 'Mahavir','Narela', 'Malviya Nagar','Vasundhara','Sheikh','Kalkaji','KalkaJi','Malviya',
       'Chittaranjan' ]
    for i in lis:
        if i in splits:
            return i
    return x
correct('Lajpat Nagar 1')
```

```python
df.Locality.apply(lambda x:x if x in loc else 'others')
df.Locality.unique()
```

```python
df.Locality=df.Locality.apply(correct)
```

```python
df.Locality.unique()
```

```python
dummies=pd.get_dummies(df.Locality)
finaldf=pd.concat([df,dummies],axis=1)
finaldf
```

```python
x=finaldf.drop(["Price","Locality"],axis=1)
y=finaldf.Price
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y)
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
mod1=LinearRegression()
mod2=Ridge()
mod3=Lasso()
```

```python
mod1.fit(xtrain,ytrain)
mod2.fit(xtrain,ytrain)
mod3.fit(xtrain,ytrain)
mod1.score(xtrain,ytrain),mod1.score(xtest,ytest),mod2.score(xtrain,ytrain),mod2.score(xtest,ytest),mod3.score(xtrain,ytrain),mod3.score(xtest,ytest)
```
